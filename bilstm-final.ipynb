```python
# Import Libraries
import pandas as pd
import numpy as np
import re
import string
import emoji
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout, SpatialDropout1D, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import matplotlib.pyplot as plt

# Configuration
CONFIG = {
    'vocab_size': 30000,
    'embed_dim': 128,
    'lstm_units': 64,
    'dense_units': 64,
    'dropout_rate': 0.5,
    'spatial_dropout_rate': 0.2,
    'l2_reg': 0.001,
    'batch_size': 128,
    'epochs': 20,
    'max_seq_len': 150,
    'test_size': 0.2,
    'random_state': 42,
    'min_learning_rate': 1e-5
}

# Function to load Sentiment140 dataset
def load_data(filepath):
    columns = ['target', 'ids', 'date', 'flag', 'user', 'text']
    df = pd.read_csv(filepath, encoding='latin-1', names=columns)
    df = df[['target', 'text']]
    df['target'] = df['target'].apply(lambda x: 0 if x == 0 else 1)  # Map 4 to 1
    return df

# Function to clean tweets
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)
    text = re.sub(r'\@\w+|\#','', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = emoji.replace_emoji(text, '')
    text = re.sub(r'\d+', '', text)
    text = text.strip()
    return text

# Function to clean and compute class weights
def preprocess_data(df):
    df['clean_text'] = df['text'].apply(clean_text)
    df = df[df['clean_text'].str.strip().astype(bool)]
    class_weights = class_weight.compute_class_weight(
        class_weight='balanced',
        classes=np.unique(df['target']),
        y=df['target']
    )
    class_weights = dict(enumerate(class_weights))
    return df, class_weights

# Main execution
if __name__ == "__main__":
    # Load and preprocess data
    df = load_data('/kaggle/input/sentiment140/training.1600000.processed.noemoticon.csv')
    df, class_weights = preprocess_data(df)

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        df['clean_text'].values,
        df['target'].values,
        test_size=CONFIG['test_size'],
        random_state=CONFIG['random_state'],
        stratify=df['target']
    )

    # Tokenization
    tokenizer = Tokenizer(
        num_words=CONFIG['vocab_size'],
        oov_token='<OOV>',
        lower=True,
        filters=''
    )
    tokenizer.fit_on_texts(X_train)

    # Sequence conversion and padding
    X_train_seq = tokenizer.texts_to_sequences(X_train)
    X_test_seq = tokenizer.texts_to_sequences(X_test)

    X_train_pad = pad_sequences(X_train_seq, maxlen=CONFIG['max_seq_len'], padding='post', truncating='post')
    X_test_pad = pad_sequences(X_test_seq, maxlen=CONFIG['max_seq_len'], padding='post', truncating='post')

    # Model Architecture
    model = Sequential([
        Embedding(
            input_dim=CONFIG['vocab_size'] + 1,
            output_dim=CONFIG['embed_dim'],
            input_length=CONFIG['max_seq_len'],
            name='embedding'
        ),
        SpatialDropout1D(CONFIG['spatial_dropout_rate']),
        Bidirectional(LSTM(CONFIG['lstm_units'], return_sequences=True,
                           kernel_regularizer=l2(CONFIG['l2_reg']),
                           recurrent_regularizer=l2(CONFIG['l2_reg']),
                           name='lstm_1')),
        BatchNormalization(name='bn_1'),
        Dropout(CONFIG['dropout_rate']),
        Bidirectional(LSTM(CONFIG['lstm_units'] // 2,
                           kernel_regularizer=l2(CONFIG['l2_reg']),
                           recurrent_regularizer=l2(CONFIG['l2_reg']),
                           name='lstm_2')),
        BatchNormalization(name='bn_2'),
        Dropout(CONFIG['dropout_rate']),
        Dense(CONFIG['dense_units'], activation='relu',
              kernel_regularizer=l2(CONFIG['l2_reg']), name='dense_1'),
        BatchNormalization(name='bn_3'),
        Dropout(CONFIG['dropout_rate'] / 2),
        Dense(1, activation='sigmoid', name='output')
    ])

    # Build model explicitly
    model.build(input_shape=(None, CONFIG['max_seq_len']))

    # Compile
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
    )

    # Print summary
    model.summary()

    # Callbacks
    callbacks = [
        EarlyStopping(patience=5, monitor='val_auc', mode='max', restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=CONFIG['min_learning_rate'])
    ]

    # Training
    history = model.fit(
        X_train_pad, y_train,
        epochs=CONFIG['epochs'],
        batch_size=CONFIG['batch_size'],
        validation_data=(X_test_pad, y_test),
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=1
    )

    # Plotting accuracy and loss
    plt.figure(figsize=(12, 5))
    
    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()
```
